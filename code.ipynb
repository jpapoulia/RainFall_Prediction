```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline

```

```python
numeric_features =['MinTemp', 'MaxTemp', 'WindGustSpeed','Rainfall','Sunshine', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity3pm','Humidity9am','Pressure9am', 'Pressure3pm','Temp9am', 'Temp3pm']


# Δημιουργία του αντίστοιχου numeric transformer
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])


```

3.γ. Γράψτε κώδικα για την κατάλληλη μετατροπή των κατηγορικών μεταβλητών ώστε να μπορούν να τους διαχειριστούν οι ταξινομητές που θα χρησιμοποιήσετε.

```python
categorical_features = ['WindDir9am', 'WindDir3pm', 'WindGustDir', 'Location','Cloud9am', 'Cloud3pm']


# Δημιουργία του αντίστοιχου categorical transformer
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

3.δ. Γράψτε κώδικα για την κλιμάκωση των χαρακτηριστικών, αν την θεωρείτε απαραίτητη

H κλιμάκωση γίνεται παραπάνω με τη χρήση του StandardScaler, που κάνει Z-score normalization.

3.ε. Εκτελέστε όλα τα προηγούμενα βήμα προεκπαίδευσης (χρήση μετασχηματιστών) ώστε να είναι "καθαρό" το αρχικό σύνολο για να χρησιμοποιηθεί για την εκπαίδευση των ταξινομητών.

```python
# Δημιουργία του ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

3.στ. Αφού έχετε "καθαρίσει" το αρχικό σύνολο εκπαίδευσης, γράψτε τον κατάλληλο κώδικα ώστε από το αρχικό σύνολο εκπαίδευσης να δημιουργήσετε τα σύνολα Χ και y.

💡 `Η στήλη 'RainTomorrow' είναι η τιμή που θέλουμε να προβλέψει ο ταξινομητής μας.`



```python
y = df['RainTomorrow']
X = df.drop('RainTomorrow', axis=1)
```

```python
#Χωρίζουμε τα data kai ta labels σε training (70%) και σε testing set (30%)


from sklearn.model_selection import train_test_split


X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=0)

X_train.shape, X_val.shape
```

Η μεγάλη ανισορροπία στα δεδομένα, όπου μια κλάση έχει πολύ περισσότερα παραδείγματα από μια άλλη, μπορεί να οδηγήσει σε μη ικανοποιητικά αποτελέσματα κατά την εκπαίδευση του μοντέλου. Η ισοζυγισμένη εκπαίδευση με χρήση βαρών είναι μια τεχνική που χρησιμοποιείται για την αντιμετώπιση του προβλήματος της ανισορροπίας των κλάσεων σε ένα dataset.Αυτή η προσεγγίση μπορεί να βοηθήσει στη βελτίωση της ικανότητας του μοντέλου να αντιμετωπίζει και τις δύο κλάσεις με ισότιμο τρόπο.

Καθώς είναι επιτακτικό να χρησιμοποιήσουμε μια μέθοδο για την τεχνητή εξισορρόπηση του συνόλου δεδομένων,γι'αυτό συμπεριέλαβα τα βάρη στο ερώτημα 6 όπου για την επιλογή των βέλτιστων υπερπαραμέτρων και τη βελτιστοποίηση του μοντέλου χρησιμοποιούμε την διασταυρούμενη επικύρωση με πλέγμα υπερπαραμέτρων.

4.β. Εφαρμόστε στα εκπαιδευμένα μοντέλα από το validation set μόνο τα δείγματα χωρίς την ετικέτα τους (predict).


```python
models=[model_knn,model_nb,model_lg,model_mlp,model_svc,model_dt,model_rf]
kind= ['Model KNN', 'Model NB', 'Model LG', 'Model MLP', 'Model SVC', 'Model DT', 'Model RF']
i=0
y_pred=[]
for model in models:
 y_prediction= model.predict(X_val) #μόνο τα δείγματα χωρίς την ετικέτα τους
 y_pred.append(y_prediction)
 from sklearn.metrics import classification_report
 print(f"Classification Report for {kind[i]}:")
 i+=1
 print()
 print(classification_report(y_val, y_prediction))
```


4.γ. Συγκρίνετε την έξοδο του κάθε μοντέλου σε σχέση με τις αντίστοιχες ετικέτες του validation set και αξιολογήστε την επιδοσή τους χρησιμοποιώντας το F1 score.

Παρατηρούμε ότι βελτιστοποιώντας την επίδοσή των ταξινομητών χρησιμοποιώντας αναζήτηση πλέγματος με διασταυρούμενη επικύρωση (k-fold 5) από την εύρεση των βέλτιστων υπερπαραμέτρων βρίσκουμε ότι **ο καλύτερος ταξινομητής είναι ο SVC** με καλύτερες υπερπαραμέτρους τις **{'classifier__C': 1, 'classifier__class_weight': {0: 1, 1: 2}, 'classifier__gamma': 'scale', 'classifier__kernel': 'rbf'}.**



*   '**classifier__kernel'='rbf':**
Το "rbf" στον πυρήνα ενός μοντέλου μηχανών διανυσμάτων υποστήριξης (SVM) αναφέρεται στον πυρήνα RBF (Radial Basis Function). Αυτός ο πυρήνας χρησιμοποιείται συχνά στα SVM για την ταξινόμηση και είναι κατάλληλος για προβλήματα όπου οι κλάσεις δεν είναι γραμμικά διαχωρίσιμες στο χώρο των χαρακτηριστικών.

*   **'classifier__C'=1 :** Η παράμετρος C στον ταξινομητή Support Vector Machine (SVM) είναι η παράμετρος κόστους. Αυτή η παράμετρος ελέγχει το πόσο αυστηρά θα εφαρμοστεί η ποινή για τα λάθη ταξινόμησης.Εδώ που το C=1,μπορεί να σημαίνει ότι ο SVC θα είναι πιο ανεκτικός σε λάθη ταξινόμησης.

*  **classifier__gamma'='scale' :** Η παράμετρος gamma στον ταξινομητή Support Vector Machine (SVM) ελέγχει τον συντελεστή της συνάρτησης πυρήνα. Κατά την εκπαίδευση ενός SVM με πυρήνα, ο υπολογισμός της απόστασης μεταξύ δύο δειγμάτων στον χώρο των χαρακτηριστικών γίνεται μέσω της συνάρτησης πυρήνα. Το gamma ρυθμίζει πόσο "στενά" θα είναι αυτά τα κύρια σύνολα υποστήριξης.Επιλέγοντας την τιμή **'scale'**, o αλγόριθμος θα προσαρμόσει το gamma με βάση τη διακύμανση των χαρακτηριστικών, προσαρμόζοντας έτσι το μοντέλο στην κλίμακα των δεδομένων.


* **'classifier__class_weight': {0: 1, 1: 2}** Η παράμετρος classifier__class_weight με την τιμή {0: 1, 1: 2} σημαίνει ότι δίνεται διαφορετικό βάρος στις δύο κλάσεις του προβλήματος κατηγοριοποίησης.Η κλάση 1 έχει βάρος 2, ενώ η κλάση 0 έχει βάρος 1.Με την επιλογή αυτής της υπερπαραμέτρου αντιμετωπίζεται το πρόβλημα της ανισσοροπίας κλάσης που αναφέραμε παραπάνω.





**Συμπεράσματα**

Το F1 score είναι ένας μετρικός δείκτης που συνδυάζει την ακρίβεια (precision) και την ανάκληση (recall) σε έναν αριθμό. Χρησιμοποιείται ευρέως στο πεδίο της ταξινόμησης, ειδικότερα όταν οι κλάσεις είναι ανισορροπημένες.

Ορίζεται ως εξής:


F1=
2
⋅
(precision
⋅
recall)/
precision
+
recall



Όπου:
Η ακρίβεια (precision) είναι ο αριθμός των σωστών θετικών προβλέψεων δια τον συνολικό αριθμό των θετικών προβλέψεων.
Η ανάκληση (recall) είναι ο αριθμός των σωστών θετικών προβλέψεων δια τον συνολικό αριθμό των πραγματικών θετικών.
Το F1 score βρίσκεται στο εύρος [0, 1], όπου 1 σημαίνει τέλεια ακρίβεια και ανάκληση, ενώ το 0 σημαίνει ότι το μοντέλο δεν έκανε καθόλου σωστές προβλέψεις.

Στο πρώτο κομμάτι δουλέψαμε με τις default παραμέτρους κάθε ταξινομητή και λάβαμε ένα χαμηλότερο f1-score σε σύγκριση με το δεύτερο κομμάτι που εργαστήκαμε με τις καλύτερες υπερπαραμέτρους καθώς μεσα σ'αυτές ήταν και τα βάρη που αντιμετώπισαν ως ένα βαθμό το πρόβλημα της ανισσοροπίας του dataset.
